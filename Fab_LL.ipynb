```python
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import requests
import requests
import os
from datetime import datetime
from utils import *
import pyarrow
```

```python
URL = 'https://fabtcg.com/living-legend/'
```

```python
if arquivo_e_atual('living_legends_fabtcg.csv'):
    print("O arquivo esta atualizado.")
else:
    print("O arquivo não existe ou esta desatualizado, corrigindo.")
    apagar_arquivo('living_legends_fabtcg.csv')
    apagar_arquivo('living_legends_fabtcg.parquet')
    
    session = requests.Session()
    session.headers.update(headers)
    respose = session.get(URL)
    soup = BeautifulSoup(respose.content, 'html.parser')
    tables = soup.find_all('table', class_ = 'has-fixed-layout')
    table = tables[2]
    trs = tables[2].find_all('tr')
    herois = []
    pontos = []
    for linha in trs[1:]:
        tr = linha.find_all('td')
        herois.append(tr[1].text)
        pontos.append(tr[2].text)
    linhas = {}
    linhas['Heroi'] = herois
    linhas['Pontos'] =pontos
    df = pd.DataFrame(linhas)
    
    df.to_csv('living_legends_fabtcg.csv')
    df.to_parquet('living_legends_fabtcg.parquet')
```
